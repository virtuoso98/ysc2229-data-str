```python
class Queue:
    def __init__(self, size):
        self.items = [None for i in range(size)]
        self.tail = 0
        self.head = 0
        self.cache = 0
        self.size = size
    
    def enqueue(self, data):
        self.cache += 1
        if self.cache > self.size:
            print("overflow")
            return
        self.items[self.tail] = data
        self.tail = (self.tail + 1) % self.size
        
        
    def dequeue(self):
        self.cache -= 1
        if self.cache < self.size:
            print("underflow")
            return
        data = self.items[self.head]
        self.head = (self.head + 1) % self.size
        return data
        
```

```python
class Stack:
    def __init__(self, size):
        self.items = []
        self.size = size
        self.cache = 0
    
    def isEmpty(self):
        return self.items == []

    def push(self, item):
        self.cache += 1
        if self.cache > self.size:
            raise ValueError("overflow")        
        self.items.append(item)
    
    def pop(self):
        self.cache -= 1    
        if self.cache < 0:
            raise ValueError("underflow")   
        self.items.pop()
    
    def size(self):
        return self.size
    
    def cache(self):
        return self.cache
    
```
